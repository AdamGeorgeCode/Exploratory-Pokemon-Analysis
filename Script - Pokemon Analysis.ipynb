### Q. How many Pokemon are there?

```python
df.shape
```

A. There are 801 Pokemon.

### Q. What generation has the most Pokemon?

```python
gencount = df.pivot_table( index = 'generation', aggfunc='count')
gencount
```

```python
colist = list(gencount.columns)
colist[1:]
```

```python
gencount = gencount.drop(labels=colist[1:], axis = 1)
```

```python
gencount.rename(columns = {'abilities':'total_pokemon'}, inplace=True)
```

```python
gencount
```

```python
gencount.sort_values(by='total_pokemon', ascending=False)
```

A. The generation that has the most pokemon is generation 5.

### Q. What is the distribution of Pokemon between the generations?

```python
gencount.plot(kind = 'bar', y= 'total_pokemon', ylabel = 'Total Pokemon', title = 'Total Pokemon Per Generation')
```

### Q. How many different Pokemon types are there?

```python
typecount = df.pivot_table( index = 'type1', aggfunc = 'count').sort_values( by = 'abilities', ascending = False)
typecount
```

```python
typecount.shape
```

A. There are 18 different Pokemom types.

### Q. What is the most common type 1 Pokemon?

```python
typecountcollist = list(typecount.columns)
typecountcollist[1:]
```

```python
typecount.drop(labels=typecountcollist[1:], axis = 1, inplace= True)
typecount
```

```python
typecount.rename(columns={'abilities':'total_pokemon'}, inplace = True)
typecount
```

A. The most common type 1 Pokemon is water.

### Q. What is the most common type 2 Pokemon?

```python
type2count = df.pivot_table( index = 'type2', aggfunc = 'count').sort_values( by = 'abilities', ascending = False)
type2count
```

```python
type2countcollist = list(type2count.columns)
type2countcollist[1:]
```

```python
type2count.drop(labels=type2countcollist[1:], axis = 1, inplace= True)
type2count
```

```python
type2count.rename(columns={'abilities':'total_pokemon'}, inplace = True)
type2count
```

A. The most common type 2 Pokemon is flying.

### Q. Across both type 1 and type 2, what is the most common Pokemon type?

```python
combinetypes = typecount.merge(type2count, left_index=True, right_index=True, how='outer')
combinetypes
```

```python
combinetypes['total_mentions'] = combinetypes['total_pokemon_x'] + combinetypes['total_pokemon_y']
combinetypes
```

```python
combinetypes.drop(columns=['total_pokemon_x','total_pokemon_y'], inplace = True)
```

```python
combinetypes.sort_values(by='total_mentions', ascending=False)
```

A. The most common type across both type 1 and type 2 is water.

### Q. What is the lowest kg Pokemon?

```python
df['weightrank'] = df['weight_kg'].rank(ascending=True, method='max')
```

```python
smallmask = df['weightrank'].min()
```

```python
df[df['weightrank'] == smallmask]
```

A. The lowest kg weighted Pokemon are Gastly, Haunter, Flabébé, Cosmog, Kartana.

### Q. What is the highest kg Pokemon?

```python
largemask = df['weightrank'].max()
```

```python
df[df['weightrank'] == largemask]
```

A. The highest kg weighted Pokemon are Cosmoem and Celesteela.
